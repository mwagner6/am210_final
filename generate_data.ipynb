```python
import numpy as np
import pandas as pd
```

```python
# non-faithful data
np.random.seed(210)

n = 500
error_var = 1.0

X1 = np.random.normal(0, np.sqrt(error_var), n)
X2 = -1.0 * X1 + np.random.normal(0, np.sqrt(error_var), n)
X3 = 1.0 * X1 + 1.0 * X2 + np.random.normal(0, np.sqrt(error_var), n)

# Create DataFrame
data = pd.DataFrame({
    'X1': X1,
    'X2': X2,
    'X3': X3
})

# Save to CSV
output_path = 'nonfaithful_data.csv'
data.to_csv(output_path, index=False)
```

```python
# non-gaussian data
np.random.seed(210)
n = 500

X1 = np.random.laplace(loc=0, scale=1, size=500)
X2 = 2.0 * X1 + np.random.laplace(loc=0, scale=2, size=500)
X3 = 5.0 * X1 + np.random.laplace(loc=0, scale=5, size=500)
X4 = 2.5 * X1 + 5.0 * X2 + np.random.laplace(loc=0, scale=5, size=500)

data = pd.DataFrame({
    'X1': X1,
    'X2': X2,
    'X3': X3,
    'X4': X4
})

output_path = 'nongaussian_data.csv'
data.to_csv(output_path, index=False)
```

```python
# partial-gaussian data
np.random.seed(210)
n = 500

X1 = np.random.normal(0, 1, size=500)
X2 = 2.0 * X1 + np.random.normal(0, 2, size=500)
X3 = 5.0 * X1 + np.random.laplace(loc=0, scale=5, size=500)
X4 = 2.5 * X1 + 5.0 * X2 + np.random.laplace(loc=0, scale=5, size=500)

data = pd.DataFrame({
    'X1': X1,
    'X2': X2,
    'X3': X3,
    'X4': X4
})

output_path = 'partial_gaussian_data.csv'
data.to_csv(output_path, index=False)
```

```python
# Unknown scale gaussian data
np.random.seed(210)
n = 500

X1 = np.random.normal(0, 1, 500)
X2 = 2.0 * X1 + np.random.normal(0, 2, 500)
X3 = 5.0 * X1 + np.random.normal(0, 5, 500)
X4 = 2.5 * X1 + 5.0 * X2 + np.random.normal(0, 8, 500)

data = pd.DataFrame({
    'X1': X1,
    'X2': X2,
    'X3': X3,
    'X4': X4
})

output_path = 'nonscaled_gaussian_data.csv'
data.to_csv(output_path, index=False)
```
